```python
import numpy as np
import pandas as pd
from Model_ComputeMetrics import getEvaluationMetrics
import Data_Processing as dp
from sklearn.dummy import DummyClassifier
```

```python
# get datasets
train, test = dp.importData()
train, test, X_train, X_test = dp.filterData(train, test)
```

```python
labels_train_encoded, labels_encoded_test = dp.initalizeLabels(train, test) # encode labels
```

```python
# set up dummy classifier 

clf = DummyClassifier(strategy="uniform")
clf.fit(X_train, labels_train_encoded)
train_predicted_labels = clf.predict(X_train)
```

```python
# predict on the test set
predictions = clf.predict(X_test)
```

```python
# do some evaluation on the training set

getEvaluationMetrics(np.array(labels_train_encoded), train_predicted_labels, 'dummy ', 'Training')
```

```python
# do some evaluation on the test set

getEvaluationMetrics(np.array(labels_encoded_test), predictions, 'dummy ', 'Test')
```
